```python
import pandas as pd
```

```python
prefix = "data/original_datasets/"

raw_MB = pd.read_csv(prefix + "dataset_Metabody.csv")
raw_ME = pd.read_csv(prefix + "dataset_MetaEducation.csv")
raw_MI = pd.read_csv(prefix + "dataset_MetaImagery.csv")
raw_MM = pd.read_csv(prefix + "dataset_MoveMe.csv")
```

```python
# Processing MB dataset:

#print("Info raw_MB:\n", raw_MB.info(), "\n")

df_dict = {"MB" : raw_MB , "ME" : raw_ME, "MI" : raw_MI, "MM" : raw_MM}

for name, df in df_dict.items():

  bambini_2013 = list()
  canal_2022 = list()
  bambini_2024 = list()
  lago_2024 = list()

  for idx, row in df.iterrows():

    if row["Bambini et al. (2013)"] == "Y":
      bambini_2013.append(row)
      continue

    if row["Canal et al. (2022)"] == "Y":
      canal_2022.append(row)
      continue

    if  row["Bambini et al. (2024)"] == "Y":
      bambini_2024.append(row)
      continue

    if row["Lago et al. (2024)"] == "Y":
      lago_2024.append(row)
      continue


  print(name + ": \n",
    "bambini_2013: ", len(bambini_2013), "\n",
    "canal_2022: ", len(canal_2022), "\n",
    "bambini_2024: ", len(bambini_2024), "\n",
    "lago_2024: ", len(lago_2024), "\n"
    )



MB = raw_MB[["Metaphor", "Met_structure", "FamMetabody_Met", "SensMetabody_Met", "BodyMetabody_Met"]]
#print("Head MB:\n", MB.head())
```
